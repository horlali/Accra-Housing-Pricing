### Cleaning the price column

```python
data['price'] = data['price'].str.replace(',', '')
data['price'] = data['price'].str.replace('month', '')
data['price'] = data['price'].str.replace('day','')
data['price'] = data['price'].str.replace(' ', '')
data['price'] = data['price'].str.replace('GH₵', '')
data['price'] = data['price'].str.replace('\n/','')
#data['price'] = data['price'].astype(int)
```

Dropping rows without property name and price

```python
data.dropna(subset=['price','property_name'],inplace=True)
```

```python
data['price'] = data['price'].astype(int)
```
